11. **Método to_datetime.** La función pd.to_datetime de la biblioteca de pandas se utiliza para convertir valores en una serie (o columna) de un DataFrame en objetos de fecha y hora. El parámetro errors se utiliza para controlar cómo se manejan los errores que puedan surgir durante la conversión. 14. **Método drop.** Si queremos eliminar la columna de un dataFrame en ese caso lo que podemos hacer es utilizar el método drop. En este caso lo ocupamos con los parámetros nombre de columna, axis e inplace. 

- `axis` Este parámetro especifica si se eliminará una fila o una columna y puede tener dos valores posibles:
  - 0, elimina filas por defecto;
  - 1, eliminará columnas. 
- `inplace` Este parámetro es un booleano que controla si la operación se realizará directamente en el DataFrame original o si se creará un nuevo DataFrame sin la fila o columna que se va a eliminar. 
  - `inplace=True:` Modifica el DataFrame original directamente. La columna o fila se elimina del DataFrame original, y no se crea un nuevo DataFrame
  - `inplace=False` (predeterminado): Crea un nuevo DataFrame sin la columna o fila que se va a eliminar, y el DataFrame original no se modifica.

```python
df_covid_vz.drop('FECHA_ACTUALIZACION', axis=1, inplace=True)
df_covid_vz.drop('ID_REGISTRO', axis=1, inplace=True)
df_covid_vz.drop('ENTIDAD_RES', axis=1, inplace=True)
df_covid_vz.drop('ENTIDAD_NAC', axis=1, inplace=True)
df_covid_vz.drop('ENTIDAD_UM', axis=1, inplace=True)
df_covid_vz.drop('ORIGEN', axis=1, inplace=True)
df_covid_vz.head(5)
```

18. Definimos la variable df_top5, que guardará dentro el resultado de hacer un top con la mayor cantidad de tipos de municipios. Le decimos que únicamente con los valores de los primeros 5:

```python
# Supongamos que ya tienes la serie con los 5 valores más comunes
serie_top5 = df_covid_vz['MUNICIPIO_RES'].value_counts()[:5]

# Convierte la serie en un DataFrame para tener una columna "Municipio" y una columna "Cantidad"
df_top5 = serie_top5.reset_index()
df_top5.columns = ['Municipio', 'Cantidad']

# Crea una figura de Matplotlib
plt.figure(figsize=(10, 6))

# Crea un gráfico de barras utilizando Seaborn
sns.set(style="whitegrid")
sns.barplot(x='Municipio', y='Cantidad', data=df_top5, palette="viridis")

# Personaliza la gráfica
plt.title("Top 5 Municipios")
plt.xlabel("Municipio")
plt.ylabel("Cantidad de Registros")

# Muestra la gráfica
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Método Merge**

En el siguiente código:

- Merge combina los DataFrames df_2 y df_1 utilizando la columna CLAVE_MUNICIPIO en df_1 y MUNICIPIO_RES en df_2.
- El argumento how='left' asegura que todos los valores en df_2 se mantendrán y se agregarán los valores correspondientes de df_1 si existen.
- Luego, puedes renombrar la columna resultante a 'MUNICIPIO_RES' si lo deseas, ya que en df_merged ahora contiene los valores de MUNICIPIO correspondientes a CLAVE_MUNICIPIO en MUNICIPIO_RES.
